#### We must read the data again because for regularization, our data must be normalized/ standardized. Since, we had already dummy coded the categorical variables (meaning they were converted to numerical variables), they will also be normalized, which is undesirable.Therefore, we will seperate the X into numeric and categroical types and perform normalization on the numeric type and dummy coding (on hot encoding) on the categorical predictors. Later, we will merge the two into one dataframe. 

Note: *For normalization, we must separate the data into train and test.* 

```python
data_df = pd.read_csv("https://raw.githubusercontent.com/dguggen/BADM211_datasets/main/GerMan.csv")
```

```python
X = data_df.drop(columns=['y', "ID"])
y = data_df['y']
```

```python
X_num_cols = X.select_dtypes(include=np.number).columns
X_num =  X[X_num_cols]
X_num.shape
```

```python
X_cat_cols = list(set(X.columns) - set(X_num_cols))
X_cat = X[X_cat_cols]
print(X_cat.shape)

X_cat = pd.get_dummies(X_cat, drop_first = True)

print(X_cat.shape)
```

```python
train_X_num, test_X_num, train_y, test_y = train_test_split(X_num, y, test_size = 0.2, random_state = 1)

train_X_cat, test_X_cat, train_y, test_y = train_test_split(X_cat, y, test_size = 0.2, random_state = 1)

print(train_X_cat.shape)
print(test_X_cat.shape)
```

#### Because all of the algorithms used in regularization require variables to be centered and on the same scale, standardize the training predictors and map that standardization to the test predictors. Be sure to recast the standardized data objects as pandas dataframes so we can use pandas functions to display those results.

***Perform this standardization only on the numeric variables, using train_X_num and test_X_num***


```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

columnnames_num =  train_X_num.columns

train_X_num = scaler.fit_transform(train_X_num)
train_X_num = pd.DataFrame(train_X_num, columns = columnnames_num)


test_X_num = scaler.transform(test_X_num)
test_X_num = pd.DataFrame(test_X_num, columns = columnnames_num)
```

```python
print(train_X_cat.shape)
print(train_X_num.shape)
print(test_X_cat.shape)
print(test_X_num.shape)
```

```python
train_X_cat.index = train_X_num.index
test_X_cat.index = test_X_num.index
```
